```python
class visualization:
    def __init__(self, df_name):
        self.df_name = df_name
        # self.x_axis = x_axis
        # self.y_axis = y_axis
        # self.plot_type = plot_type
        
    def plotter(self, module_name, plot_type, x_axis, y_axis):
        data = self.df_name
        getattr(module_name, plot_type)(x = x_axis, y = y_axis,
                                    data=data)
```

```python
data = visualization(df)
data.plotter(sns, 'swarmplot', 'species', 'petal_length' )
# data_visualization(df).plotter(sns, 'swarmplot', 'species', 'petal_length' )
```

# Scaling of plot elements

The four preset contexts, in order of relative size, are <code> paper </code>, <code>notebook</code>, <code>talk</code> and  <code>poster</code>.

```python
sns.set(style = 'whitegrid', rc={'grid.linewidth': 1.5})
sns.set_context('poster', font_scale = 2.5, rc = {'lines.linewidth': 5.0})
sinplot()
```
